## Principal Component Analysis - Dimensionality Reduction

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## load the dataset 
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df = pd.DataFrame(cancer_dataset['data'], columns=cancer_dataset['feature_names'])
df
```

## Standarization

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

## Applying PCA algorithm

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # Extracting all the features into 2 features
```

```python
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca # Now I've only 2 columns -> PCA Transform all the columns into 2 columns 
```

```python
pca.explained_variance_ # w.r.t for 2 features -> I'm getting variance
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0], data_pca[:, 1], c=cancer_dataset['target'], cmap='plasma')
plt.xlabel('First Principal Components')
plt.ylabel('Second Principal Components')
```
